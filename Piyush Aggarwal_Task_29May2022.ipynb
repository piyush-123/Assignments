```python
import time
import datetime
from datetime import datetime
scheduled_time = input('Enter the scheduled time in HH:MM:SS format')
scheduled_time = datetime.strptime(scheduled_time,"%H:%M:%S")
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
current_time = datetime.strptime(current_time,"%H:%M:%S")

print("current time : ", current_time)

if current_time > scheduled_time:
    print('Time Already Passed')
else:
    time_diff = (scheduled_time - current_time).total_seconds()
    
    print("Alarm will trigger after " , time_diff ," seconds" )
    time.sleep(time_diff)
    

    print("Alarm triggered")
    
        

```

```python
#q16 : Write a fun which will take input as a dict and give me out as a list of all the values 
#even in case of 2 level nesting it should work .

d = {'k1':1,'k2':2,'k3':3,'k4':{'k5':5},'k6':{'k7':{'k8':6.8}},'k9':7}
l = list(d.keys())

for i in range(len(l)):
    if type(d[l[i]]) == int or type(d[l[i]]) == float:
        print(d[l[i]])
    elif type(d[l[i]]) == dict:
        l1 = list(d[l[i]].keys())
        for j in range(len(l1)):
            if type(d[l[i]][l1[j]]) == int or type(d[l[i]][l1[j]]) == float:
                print(d[l[i]][l1[j]])
            elif type(d[l[i]][l1[j]]) == dict:
                        l2 = list(d[l[i]][l1[j]].keys())
                        for k in range(len(l2)):
                            print(d[l[i]][l1[j]][l2[k]])
                
```
